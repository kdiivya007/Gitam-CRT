# increment and decrement

```
#include<stdio.h>
int main()
{
	int a,b;
	a=10;
	b=0;
	printf("\n a=%d b=%d",a,b);
	b=a;
	printf("\n a=%d b=%d",a,b);
	b=++a;
	printf("\n a=%d b=%d",a,b);
	b=--a;
	printf("\n a=%d b=%d",a,b);
	b=a++;
	printf("\n a=%d b=%d",a,b);
	b=a--;
	printf("\n a=%d b=%d",a,b);
	return 0;
}
```

# switch case

```
# include<stdio.h>
int main()
{
	int x;
	scanf("%d",&x);
	switch(x)
	{
	
	 case 1:printf("monday");
	 break;
	 case 2:printf("tuesday");
	 		break;
		 
		 case 3:printf("wednesday");
		 break;
	
	 case 4:printf("thursday");
	 	break;
	 
	 case 5:printf("friday");
	 break;
	 case 6:printf("saturday");
	 break;
	 case 7:printf("sunday");
}
return 0;
}
```

# vowel consonent switch case




```
# include<stdio.h>
int main()
{
	int x;
	scanf("%c",&x);
	switch(x)
	{
		
		case 'a':
			printf("vowel");
		break;
		case 'e':
		    printf("vowel");
		break;
		case 'i':
		    printf("vowel");
		break;
		case 'o':
		    printf("vowel");
		break;
		case 'u':
		    printf("vowel");
		break;
		default :
		printf("consonent");
	}
	return 0;
}
```
