# nth posintion in a palindrome

```
#include<stdio.h>
int isPalindrome(int x)
{
	int rev=0,buffer=x;
	while(x!=0)
	{
		rev=rev=rev*10+(x%10);
		x/=10;
	
	}
	if(buffer==rev)
	{
		return 1;
	}
	else
	{
		return 0;
	}
}
int main()
{
	int cnt=0,x1,x2,n,res,i;
	scanf("%d%d",&x1,&x2);
	scanf("%d",&n);
	for(i=x1;i<=x2;i++)
	{
		res=isPalindrome(i);
		if(res==1)
		{
			cnt++;
			if(cnt==n)
			{
				printf("%d",i);
				break;
			}
		}
	}
}
```

# read the input  and print the output "yes" or "no"
- test cases
   - 145--yes
   - 123-- no

```
#include<stdio.h>
int fact(int x)
{
	int i,f=1;
	for(i=x;i>0;i--)
	{
	   f=f*i;	
	}
	return f;
}
int main()
{
	int d,r,buffer,sum=0;
	scanf("%d",&d);
	buffer=d;
	while(d!=0)
	{
		r=d%10;
		sum=sum+fact(r);
		d/=10;
	}
	if(buffer==sum)
	{
		printf("yes");
	
	}
	else
	{
		printf("no");
	}


	
	
}
```

```
#include<stdio.h>
#include<math.h>
int digCount(int a )
{
	int cnt=0;
	while(a!=0)
	{
		cnt++;
		a/=10;
	}
	return cnt;}
int main()
{
	int x,r,sum=0,count;
	scanf("%d",&x);
	count=digCount(x);
	while(x!=0)
	{
		r=x%10;
		sum=sum+pow(r,count);
			x/=10;
	}
	printf("%d",sum);

	
	return 0; 
}
```
